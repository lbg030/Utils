```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
nemo = cv2.imread('/Users/ibyeong-gwon/Desktop/Git/Utils/190522-3-4_patch0.jpg')
plt.imshow(nemo)
plt.show()
```

```python
nemo = cv2.cvtColor(nemo, cv2.COLOR_BGR2RGB)
plt.imshow(nemo)
plt.show()
```

```python
hsv_nemo = cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)
```

```python
light_orange = (77, 77, 77)
dark_orange = (128, 128, 128)
```

```python
mask = cv2.inRange(hsv_nemo, light_orange, dark_orange)
```

```python
result = cv2.bitwise_and(nemo, nemo, mask=mask)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(mask)
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
